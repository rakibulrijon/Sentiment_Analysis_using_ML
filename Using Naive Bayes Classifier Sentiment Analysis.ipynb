# Data Cleaning: Creating a pipeline with stop words , punctuation marks  and count vectorizer 

```python
import string
string.punctuation
```

```python
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
stopwords.words("english")
```

```python
def message_cleaning(Test):
    Test_punc_removed=[char for char in Test if char not in string.punctuation]
    Test_punc_removed_join="".join(Test_punc_removed)
    Test_punc_removed_join_clean=[word for word in Test_punc_removed_join.split() if word.lower() not in stopwords.words("english")]
    return Test_punc_removed_join_clean
```

```python
data_df["Comment"]= data_df["Comment"].apply(message_cleaning)
```

```python
data_df["Comment"].head()
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
def word_lemmatizer(comment):
    lem_comment=[lemmatizer.lemmatize(i) for i in comment]
    return lem_comment
```

```python
data_df["Comment"].apply(lambda x: word_lemmatizer(x))
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
def word_stemmer(comment):
    stem_comment=" ".join([stemmer.stem(i) for i in comment])
    return stem_comment
```

```python
data_df["Comment"]=data_df["Comment"].apply(lambda x:word_stemmer(x))
```

```python
data_df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(use_idf= True,lowercase= True, strip_accents=ascii,min_df=5, max_df = 0.8)
```

```python
X=vectorizer.fit_transform(data_df["Comment"])
```

```python
y=data_df["Sentiments"]
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test= train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
NB_classifier = MultinomialNB()
NB_classifier.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_predict_test=NB_classifier.predict(X_test)
cm=confusion_matrix(y_test,y_predict_test)
sns.heatmap(cm,annot=True)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test,y_predict_test))
```

```python
review=np.array(["This is an excellent album with some great fiddle playing. Ryan is amazing to watch when he plays in person. You can feel the emotion he puts into each song."])
review_vector=vectorizer.transform(review)
x= NB_classifier.predict(review_vector)
```
